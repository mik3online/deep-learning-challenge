## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value
cutoff = 500

# Create a list of application types to be replaced
application_types_to_replace = application_type_counts[application_type_counts < cutoff].index.tolist()

# Replace in the DataFrame
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# Look at CLASSIFICATION value counts > 1
classification_counts_gt1 = classification_counts[classification_counts > 1]
print(classification_counts_gt1)
```

```python
# Choose a cutoff value
cutoff = 1000

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts_gt1[classification_counts_gt1 < cutoff].index.tolist()

# Replace in the DataFrame
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# Convert categorical data to numeric with pd.get_dummies
application_df = pd.get_dummies(application_df)
```

```python
# Split the preprocessed data into features (X) and target (y) arrays
X = application_df.drop(columns='IS_SUCCESSFUL')
y = application_df['IS_SUCCESSFUL']

# Split the data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

**bold text**## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=len(X_train.columns)))

# Second hidden layer (you can add this layer if needed)
nn.add(tf.keras.layers.Dense(units=30, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Define the number of epochs
num_epochs = 100

# Lists to store training history
loss_history = []
accuracy_history = []

# Training loop
for epoch in range(1, num_epochs + 1):
    history = nn.fit(X_train, y_train, epochs=1, verbose=2)
    loss_history.append(history.history['loss'][0])
    accuracy_history.append(history.history['accuracy'][0])

    # Print the current epoch's progress
    print(f'Epoch {epoch}/{num_epochs}')
    print(f'Loss: {history.history["loss"][0]:.4f} - Accuracy: {history.history["accuracy"][0]:.4f}')

# Display the training history
for epoch in range(num_epochs):
    print(f'Epoch {epoch + 1}/{num_epochs}')
    print(f'Loss: {loss_history[epoch]:.4f} - Accuracy: {accuracy_history[epoch]:.4f}')
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
